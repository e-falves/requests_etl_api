```python
#Importando biblioteca Requests
import requests
```

```python
#Primeira versão Headers
headers = {'X-GitHub-Api-Version': '2022-11-28'}
```

```python
#Construção da url da api Github
api_base_url = 'https://api.github.com'
owner = 'amzn'
url = f'{api_base_url}/users/{owner}/repos'

url
```

```python
#Acessando a api
response = requests.get(url, headers=headers)

response.status_code
```

```python
#Validando o tamanho dos dados
len(response.json())
```

```python
#Segunda versão do Headers com tokem de acesso
access_token = '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'
headers = {'Authorization': 'Bearer' + access_token,
           'X-GitHub-Api-Version': '2022-11-28'}
```

```python
#Extraindo repositórios
repos_list = []

for page_num in range(1,7):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

```python
#Conhecendo os dados
repos_list[0][0]
```

```python
#Separando os repositórios e as linguagens utilizadas
list_name = {'Repositorio':[],
             'Linguagem':[]}

for page in repos_list:
    for repos in page:
        repositorio = repos['name']
        linguagem = repos['language']
        list_name['Repositorio'].append(repositorio)
        if linguagem == None:
            list_name['Linguagem'].append('Não Definido no repositório')
        else:
            list_name['Linguagem'].append(linguagem)
```

```python
#Conhecendo os dados
list_name
```

```python
#Validando o tamanho dos dados
len(list_name['Linguagem'])
```

```python
#Importando biblioteca Pandas
import pandas as pd

#Transformando em DataFrame
dados_amzn = pd.DataFrame(data=list_name)
dados_amzn
```
